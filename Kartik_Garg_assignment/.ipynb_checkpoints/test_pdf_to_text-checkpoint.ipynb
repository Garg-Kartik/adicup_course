```python
#imports be careful because there are lot of pdf miner
from pdfminer.converter import TextConverter
from pdfminer.pdfinterp import PDFResourceManager
from pdfminer.pdfinterp import PDFPageInterpreter
from pdfminer.pdfpage import PDFPage
from pdfminer.layout import LAParams
from cStringIO import StringIO

pdf_file = open('beaches.pdf','rb')
manager = PDFResourceManager()
layout = LAParams()
result = StringIO()
device = TextConverter(manager,result,laparams=layout)
interpreter = PDFPageInterpreter(manager,device)

#process the file
pages = {}
npages = 0
nwords = 0
words={}
for page in PDFPage.get_pages(pdf_file):
    interpreter.process_page(page) #pdfMiner --> StringIO
    page_data = result.getvalue() #StringIO --> str
    #store individual words
    for word in page_data.split():
        if word in words.keys():
            words[word] += 1 #increment count
        else:
            words[word] = 1 #initialize
    nwords = nwords + len(page_data.split())
    pages[npages] = page_data
    npages = npages + 1
    result.truncate(0) #reset

print("Number of words = " +str(nwords))
for i in range(npages):
    print pages[i]
pdf_file.close()
device.close()
```

```python
dictlist=[]
for key, value in pages.iteritems():
    temp = [key,value]
    dictlist.append(temp)
# print(dictlist)
list1=reduce(lambda x,y :x+y ,dictlist)
# list1
str1 = ''.join(str(e) for e in list1)
list2=str1.split()
from collections import Counter
counts = Counter(list2)
print(counts)
```
