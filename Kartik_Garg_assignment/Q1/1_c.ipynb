```python
import pandas as pd 
%matplotlib notebook
import matplotlib.pyplot as plt


df=pd.read_csv("ncdc_data.csv")
# df['t']=(df['YEAR']-1990)*365+df['DAY']
Year=[]
temp=[]
index=[]

temps=df['TEMP'].tolist()
Days=df['DAY'].tolist()
Years=df['YEAR'].tolist()

for year in Years:
    if year not in Year:
        Year.append(year)
for year in Year:
    index=df.index[df['YEAR']==year].tolist()
    for i in index:
        if i==1822:
            break
        else:
            temp1=df.loc[i]['TEMP']
            temp2=df.loc[i+1]['TEMP']
            temp3=df.loc[i+2]['TEMP']
            temp.append(temp1+temp2+temp3)
    index[:]=[]
```

```python
import matplotlib.pyplot as plt 
plt.plot( temp, color='black', linewidth = 2, marker='o', markerfacecolor='red', markersize=4) 
  
# setting x and y axis range 
plt.ylim(min(temp),max(temp)) 
plt.xlim(min(temp),max(temp)) 
plt.xlabel('Time') 
# naming the y axis 
plt.ylabel('Heat wave') 
  
# giving a title to my graph 
plt.title('Heat Wave vs Time') 
  
# function to show the plot 
plt.show()
```
