```python
from matplotlib import pyplot as plt
import random

# some sports data - point differentials among AFC East teams
# in the NFL over past 25 years
afc = { 'Years':[2017,2016,2015,2014,2013,2012,2011,2010,2009,2008,\
                 2007,2006,2005,2004,2003,2002,2001,2000,1999,1998,\
                 1997,1996,1995,1994,1993],\
        'Bills':[  -57,   21,   20,   54,  -49,  -91,  -62, -142, -68,\
                   -6,  -102,  -11,  -96,  111,  -36,  -18, -155, -35,\
                   91,    67, -112,   53,   15,  -16,   87],\
        'Patriots':[162, 191,  150,  155,  106,  226,  171,  205,  142,\
                    101, 315,  148,   41,  177,  110,   35,   99,  -62,\
                     15,   8,   80,  105,  -83,   39,  -48],\
        'Jets':[  -84,  -134,   73, -118,  -97,  -94,   14,   63,  112,\
                   49,   -87,   21, -115,   72,  -16,   23,   13,    0,\
                   -1,   150,   61, -175, -151,  -56,   23],\
        'Dolphins':[-112,-17,  -79,   15,  -18,  -29,   16,  -60,  -30,\
                     28,-170,  -23,    1,  -79,   50,   77,   54,   97,\
                    -10,  56,   12,   14,   66,   62,   -2]}

# extract data into lists
xnames = [ 'Bills','Patriots','Jets','Dolphins' ]
bills = afc['Bills']; pats = afc['Patriots']
jets = afc['Jets']; fins = afc['Dolphins']

# list of lists for pyplot
box_data = [bills,pats,jets,fins]
# jittered x-coords
eps = 0.1 # control amount of jitter
xb = [random.uniform(1-eps,1+eps) for i in range(0,25)]
xp = [random.uniform(2-eps,2+eps) for i in range(0,25)]
xj = [random.uniform(3-eps,3+eps) for i in range(0,25)]
xf = [random.uniform(4-eps,4+eps) for i in range(0,25)]
# pipeline
plt.plot(xb,bills,'ro')
plt.plot(xp,pats,'bo')
plt.plot(xj,jets,'go')
plt.plot(xf,fins,'co')
plt.boxplot(box_data, labels=xnames, sym='')
plt.title("AFC East Point Differentials")
plt.xlabel("Team")
plt.ylabel("Differential")
#plt.savefig("box.png", dpi=100)
plt.show()
plt.close()


```
