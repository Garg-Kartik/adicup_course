```python
#imports be careful because there are lot of pdf miner
from pdfminer.converter import TextConverter
from pdfminer.pdfinterp import PDFResourceManager
from pdfminer.pdfinterp import PDFPageInterpreter
from pdfminer.pdfpage import PDFPage
from pdfminer.layout import LAParams
from cStringIO import StringIO

def files(file1):
    pdf_file = open(file1,'rb')
    manager = PDFResourceManager()
    layout = LAParams()
    result = StringIO()
    device = TextConverter(manager,result,laparams=layout)
    interpreter = PDFPageInterpreter(manager,device)

    #process the file
    pages = {}
    npages = 0
    nwords = 0
    words={}
    for page in PDFPage.get_pages(pdf_file):
        interpreter.process_page(page) #pdfMiner --> StringIO
        page_data = result.getvalue() #StringIO --> str
        #store individual words
        for word in page_data.split():
            if word in words.keys():
                words[word] += 1 #increment count
            else:
                words[word] = 1 #initialize
        nwords = nwords + len(page_data.split())
        pages[npages] = page_data
        npages = npages + 1
        result.truncate(0) #reset

    print("Number of words = " +str(nwords))
#     for i in range(npages):
#          print pages[i]
    pdf_file.close()
    device.close()
    return pages
```

```python
def convert(pages):
    dictlist=[]
    for key, value in pages.iteritems():
        temp = [key,value]
        dictlist.append(temp) #converto to list
    list1=reduce(lambda x,y :x+y ,dictlist) #convert to 1d list
    str1 = ''.join(str(e) for e in list1)  #convert list to string
    list2=str1.split() #cleaned list
    return list2
```

```python
file1_pages=files('beaches.pdf')
file2_pages=files('gettysburg.pdf')
file3_pages=files('pearl_harbor.pdf')
```

```python
file1_list=convert(file1_pages)
file2_list=convert(file2_pages)
file3_list=convert(file3_pages)
```

```python
def intersection(lst1, lst2, lst3): 
    return list(set(lst1) & set(lst2) &set(lst3)) 
```

```python
intersection_list=intersection(file1_list,file2_list,file3_list)
```

```python
def count(list1): 
    from collections import Counter
    counts = Counter(list1)
    return counts
#     print(counts)
```

```python
file1_counts=count(file1_list)
file2_counts=count(file2_list)
file3_counts=count(file3_list)

intersection_counts= count(intersection_list)
```

```python
print("--------Words and their iterations in : *beaches.pdf*-------------")
for key, value in file1_counts.most_common():
    print(key, value)
```

```python
print("---------Words and their iterations in : *gettysburg.pdf*-------------")
for key, value in file2_counts.most_common():
    print(key, value)
```

```python
print("---------Words and their iterations in : *pearl_harbor.pdf*-------------")
for key, value in file3_counts.most_common():
    print(key, value)
```

```python
print("---------Words and their iterations common in all three pdf's-------------")
for key, value in intersection_counts.most_common():
    print(key, value)
```
